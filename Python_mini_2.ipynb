```python
columns=[
    'Sex','Length','Diameter','Height',
'Whole weight','Shucked weight','Viscera weight',
    'Shell weight','Rings'
]

df=pd.read_csv('abalone.data', header=None, names=columns)

df.head()
```

Load the dataset and display the first 5 rows.

```python
df.head()
```

Check for missing values in each column


```python
df.isnull().sum()
```

Display summary statistics for all numerical columns

```python
df.describe()
```

Count the number of occurrences for each category in the 'Sex' column.

```python
df['Sex'].value_counts()
```

Calculate the average 'Length' for each category in 'Sex'.

```python
df.groupby('Sex')['Length'].mean()
```

Create a new column 'Age' by adding 1.5 to the 'Rings' column.

```python
df['age']=df['Rings']+1.5
```

```python
df.head()
```

Create a new column 'TotalWeight' by summing all weight-related columns

```python
df['TotalWeight']=df['Whole weight']+df['Shucked weight']+df['Viscera weight']+df['Shell weight']
df.head()
```

Filter and display rows where 'Height' is greater than 0.2.

```python
filtered_df=df[df['Height']>0.2]
print(filtered_df)
```

```python
print(df[df['Height']>0.2])
```

Sort the dataset in descending order based on 'Rings'.

```python
desc_order=df.sort_values('Rings',ascending =False).head()
print(desc_order)
```

Group records into age brackets and count the number of records in each bracket.

```python
age_counts=df['age'].value_counts().sort_index()
print(age_counts)
```
